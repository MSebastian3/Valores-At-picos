```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('ventastotales_sin_nulos.csv') 
```

```python
#Obtengo información del Dataframe
data = data.drop('Unnamed: 0', axis=1)
data.head()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

## **ventas_precios_corrientes**

```python
#Realizamos histograma de variables cuantitativas
fig = plt.figure(figsize =(5, 3))
plt.hist(x=data["ventas_precios_corrientes"], color='blue', rwidth=0.50)
plt.title('Histograma de Precio de ventas corrientes')
plt.xlabel('Precio Venta')
plt.ylabel('Frecuencia')

plt.xticks(rotation=45)

plt.show() #dibujamos el histograma
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(10, 6))
plt.boxplot(data["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Venta")
plt.show() #dibujamos el diagrama
```

**Método de desviación estándar**



```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean1= data[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean1
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean1["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Venta")
plt.show() #dibujamos el diagrama
```

**MÉTODO DE IQR (RANGO INTER QUARTILICO)**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr1= data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr1
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr1["ventas_precios_corrientes"]) 
plt.title("Outliers de Precio Venta")
plt.show() #dibujamos el diagrama
```

```python
#Convertir archivo filtrado a CSV
data_clean_iqr1.to_csv("ventas_precios_corrientes.csv")

from google.colab import files 
files.download("ventas_precios_corrientes.csv")
```

Para la columna de ventas_precios_corrientes se encontró un valor outlier con el metodo de Rango Inter Cuartílico, con el método de desviación estándar no, debido a que el metodo es mas tolerante.
