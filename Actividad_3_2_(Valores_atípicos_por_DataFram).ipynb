# 2020

```python
#Carga desde un archivo .csv sin indice
df2020= pd.read_excel('G&C_LIMPIA.xlsx', sheet_name="2020")
df2020.info()
```

```python
#df2020.head()
```

```python
#Corroboramos valores nulos
df2020["FOLIO"]=df2020["FOLIO"].fillna(00000)
valores_nulos=df2020.isnull().sum()
valores_nulos
```

```python
#filtro no numericas
filtro2020=df2020.iloc[ : , [0,1,2,3,4,5,6,7,15,16,17,18]]
#print(filtro2020numericas.head())

#filtro numericas
filtro2020numericas=df2020.iloc[ : , [8,9,10,11,12,13,14]]
#filtro32020no.head()
```

## Procedimiento

## **PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=filtro2020numericas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
data2= y.copy()
outliers= data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Eliminamos columnas innecesarias
#data4= data3.copy().drop(['Columna1'], axis=1)
#data4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=filtro2020numericas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Eliminamos columnas innecesarias
#data4_iqr= data3_iqr.copy()#.drop(['indice_tiempo'], axis=1)
#data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=data3_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data3_iqr.mean(),1))
data5_iqr
```

```python
# Unimos la columna 'indice_tiempo' con el dataframe

#indice_tiempo=data2['indice_tiempo']
Datos_limpios = data5_iqr.copy()#pd.concat([indice_tiempo, data5_iqr], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

## Concatenar/Guardar

```python
concatenadodesv = pd.concat([filtro2020, data_clean], axis=1)
concatenadodesv.head()
```

```python
concatenadoquan = pd.concat([filtro2020, Datos_limpios], axis=1)
concatenadoquan.head()
```

```python
#Convertir DataFrame a CSV
concatenadodesv.to_csv("2020_1_limpio.csv")
concatenadoquan.to_csv("2020_2_limpio.csv")
```

```python
#descargar archivo filtrado en csv
from google.colab import files 
  
files.download("2020_1_limpio.csv")
files.download("2020_2_limpio.csv")
```
